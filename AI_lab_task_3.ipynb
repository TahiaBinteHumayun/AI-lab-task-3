```python
course_list = []
option = 0

def line():
    print('__________________________________________________')


def dashboard():
    line()
    print('Dashboard')
    line()
    print('1. Add course')
    print('2. Remove course')
    print('3. Taken courses')
    print('4. Logout')
    line()
    option = input('Choose option: ')
    return option 


while option != '4':
    option = dashboard()
    if option == '1':
        course = input('Course name you want to enroll in? ')
        course_list.append(course)
        print('You have successfully enrolled in ', course, 'course.\n')
    elif option == '2':
        course = input('Which course to remove? ')
        if course in course_list:
            course_list.remove(course)
            print('Removed item', course)
        else:
            print('You are not enrolled in ', course, "course.\n")
    elif option == '3':
        print('Your enrolled courses for this semester:')
        for course in course_list:
            print(course)
    elif option == '4':
        print('Courses you are enrolled in are:', course_list)
    else:
        print('Enter the right option number.')
```
